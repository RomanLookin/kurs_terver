4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

```python
a=174
std=8
```

a) рост больше 182 см = a+std. Необходимо найти Р(Х>182 cm)=P(x>a+std)=P(X<a-std)=0.1587
б) рост больше 190 см = a+2*std. Необходимо найти Р(Х>190 cm)=P(x>a+2*std)=P(X<a-2*std)=0.0228
в) рост от 166 см до 190 см. Необходимо найти Р(Х>166 & X<190)=1-P(x>a+2*std)-P(x<a-std)=1-0.0228-0.1587=0.8185
г) рост от 166 см до 182 см. Необходимо найти Р(Х>166 & X<182)=1-P(x>a-std)-P(x<a+std)=1-2*0.1587=0.6844
д) рост от 158 см до 190 см. Необходимо найти Р(Х>158 & X<190)=1-P(x>a-2*std)-P(x<a+2*std)=1-2*0.0228=0.9544
е) рост до 150 см и от 190 см. Необходимо найти Р(Х<150 & X>190)=P(x<a-3*std)+P(x>a+2*std)=0.0228+0.0014=0.0242
ё) рост до 150 см и от 198 см. Необходимо найти Р(Х<150 & X>198)=P(x<a-3*std)+P(x>a+3*std)=0.0014+0.0014=0.0028
ж) рост ниже 166 см = a-std. Необходимо найти Р(Х<166 cm)=P(x<a-std)=0.1587
